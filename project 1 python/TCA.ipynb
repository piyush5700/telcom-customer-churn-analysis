```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By Churn")
plt.show
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn' : "count"})
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%")
plt.title("Percentage Of Churned Customers")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(data = df , x ="SeniorCitizen")
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By SeniorCitizen")
plt.show()
```

```python
# Create a cross-tab
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100

# Plot stacked bar
ax = ct.plot(kind="bar", stacked=True, figsize=(4,4), colormap="coolwarm")

# Add percentage labels
for i in range(len(ct)):
    cumulative = 0
    for j in range(len(ct.columns)):
        value = ct.iloc[i, j]
        if value > 0:
            ax.text(i, cumulative + value/2, f"{value:.1f}%", 
                    ha='center', va='center', fontsize=9)
            cumulative += value

# Labels
plt.title("Churn By SeniorCitizen (Stacked Bar Chart)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.xticks([0, 1], ["No", "Yes"], rotation=0)
plt.legend(title="Churn")

plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up subplot grid
n_cols = 3   # number of columns in grid
n_rows = (len(cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()

# Loop through columns and plot
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i],hue = df["Churn"])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#customers with fiber optic internet show noticeably higher churn than those using DSL or having no internet service.
Users without online security, online backup, and device protection are more likely to churn than users who have those services enabled.
For phone service and multiple lines, the majority of customers stay, but churn still appears higher among customers with active service than those with no service.
Overall, lack of value-added support services is strongly associated with higher customer churn.
